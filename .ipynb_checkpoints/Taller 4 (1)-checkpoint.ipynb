1 Write a function read_csv(file, missingValueSymbol) that reads in the given csv file
file and replaces all occurrences of missingValueSymbol by a NaN entry and returns
this data frame.


```python
def read_csv(file, missingValueSymbol):
    fileData= pd.read_csv(file)
    for col in fileData.columns:
        fileData[col] = fileData[col].map(lambda x: np.nan if x in [missingValueSymbol] else x)
    return fileData

#read_csv('./airlinetickets.csv', '?')
#read_csv('./baseball_numeric.csv', '?')
```

2 Write a function getNearestNeighbor(df, point) that takes in a Pandas DataFrame
and an (incomplete) data point and returns the complete data point in the data frame
that has the smallest Euclidean distance to the incomplete point (distance in the reduced
space of available features). For categorical features, assume the distance induced by
the multinomial Bernoulli representation.

```python
def getNearestNeighbor(df, point):
    D = df.copy()
    pointFormatted = point.copy()
    nanIndex = -1
    
    # Get index of the nan value in the array
    for i in range(len(point)):
        if not isinstance(point[i], str):
            if np.isnan(point[i]):
                nanIndex = i
        
    for i in range(len(df.columns)):
        col = df.columns[i]
        
        #Check categoric
        if not np.issubdtype(df[col].dtype, np.number):
            #print('is Categoric')
            indexes = df.index[df[col] != point[i]].tolist()
            exclusionCondition = df.index.isin(indexes)
            
            D.loc[df[col] == point[i], col] = 1
            D.loc[df[col] != point[i], col] = 0
            
            pointFormatted[i] = 0
    
    pointFormatted[nanIndex] = 0
    pointFormatted = pointFormatted.astype(np.float)

    zerosArray = np.zeros(D.shape[0])
    distanceDataFrame = pd.DataFrame(data=zerosArray, columns=['Distance'])

    for i, row in D.iterrows():
        secondPoint = np.array(row.values)
        secondPoint = secondPoint.astype(np.float)
        
        euclideanDistance = np.linalg.norm(pointFormatted - row.values)
        distanceDataFrame['Distance'][i] = euclideanDistance
    
    minDistance = distanceDataFrame['Distance'].min()
    
    rowMinDistance = distanceDataFrame[distanceDataFrame['Distance'] == distanceDataFrame['Distance'].min()]
    indexMinDistance = rowMinDistance.index
    
    filteredTargetValue = df.iloc[indexMinDistance, nanIndex].values[0]
    point[nanIndex] = filteredTargetValue
    
    # Insert to dataframe
    df = df.append(dict(zip(df.columns, point)), ignore_index=True)
    
    return df

# TEST 1

airlineTickets = read_csv('./airlinetickets.csv', '?')
airlineTickets.dropna(axis='columns')

# Get a point
point = np.array(airlineTickets.values[1])

# Set gender to nan value
point[4] = np.nan

# Reset indexes to apply changes
airlineTickets.drop(1, axis=0, inplace=True)
airlineTickets.reset_index(drop=True, inplace=True)

#getNearestNeighbor(airlineTickets, point)

# TEST 2
titanic = read_csv('./titanic.csv', '?')
titanic.dropna(axis='columns')

testPoint = np.array(titanic.values[1])

# Set gender to nan value
testPoint[2] = np.nan

# Reset indexes to apply changes
titanic.drop(1, axis=0, inplace=True)
titanic.reset_index(drop=True, inplace=True)

#getNearestNeighbor(titanic, testPoint)
```
