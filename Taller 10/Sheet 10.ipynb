```python
def isAContainedInAnyOfB(a,b):
    #(is A cotained or equal) in any list of B?
    a=pd.DataFrame(a).values
    b=pd.DataFrame(b).values
    return np.any(np.sum(np.isin(b,a),axis=1)>=len(a))

def isAContainedNotEqualInAnyOfB(a,b):
    #(is A cotained) in any list of B? 
    a=pd.DataFrame(a).values
    b=pd.DataFrame(b).values
    conditionA=np.sum(np.isin(b,a),axis=1)>=len(a)
    conditionB=(~np.isnan(b)).sum(1)>len(a)    
    return np.any(conditionA*conditionB)

def genMax(db,minsup):
    pset = {}
    for tIndex, t in enumerate(db):
        for i in t:
            if not i in pset:
                pset[i] = []
            pset[i].append(tIndex)
    M = []
    P = [([i],set(pset[i])) for i in sorted(pset.keys()) if len(pset[i]) >= minsup]    
    M=genMaxRec(P,minsup,M)
    return M

def genMaxRec(P,minsup,M):
    
    Y=[ x[0] for x in P]
    Y=list(set([item for sublist in Y for item in sublist]))   
    if isAContainedInAnyOfB(Y,M):        
        return
    for indexA, (Xa, tXa) in enumerate(P):         
        Pa = []
        for Xb, tXb in P[indexA+1:]:
            tXab = tXa & tXb
            if len(tXab) >= minsup:
                Pa.append((sorted(list(set(Xa + Xb))), tXab))
        if Pa:            
            genMaxRec(Pa,minsup,M)
        else:                      
            if not(isAContainedInAnyOfB(Xa,M)):
                M.append((Xa))
       
    return M

db=readTransactionalDatabase("shop.dat")
#result=eclat(db,3)
#result=[x[0] for x in result]
genMax(db,1000)
```

```python
def charm(db,minsup):
    pset = {}
    for tIndex, t in enumerate(db):
        for i in t:
            if not i in pset:
                pset[i] = []
            pset[i].append(tIndex)
    C = []
    P = [[[i],set(pset[i])] for i in sorted(pset.keys()) if len(pset[i]) >= minsup]
    C=charmRec(P,minsup,C)    
    return C
def charmRec(P,minsup,C):    
    orderP=np.array(P)     
    lens=np.argsort([ len(suports) for suports in orderP[:,1]])
    orderedTidsets=np.take_along_axis(orderP[:,1],lens, axis=0)
    orderdItems=np.take_along_axis(orderP[:,0],lens, axis=0)
    orderP=np.column_stack([orderdItems,orderedTidsets])
    for indexA, [Xa, tXa] in enumerate(orderP):
        Pi=[]
        for indexB,[Xb, tXb] in enumerate(orderP[indexA+1:]):            
            tXab = tXa & tXb            
            if len(tXab) >= minsup:
                
                if tXa==tXb :                    
                    orderP[indexA]=(list(set(Xa + Xb)),tXa)
                    del orderP[indexB]
                else:
                    
                    if isAContainedNotEqualInAnyOfB(tXa,[tXb]):
                        orderP[indexA,0]=orderP[indexA,0]+Xb                        
                        if Pi:
                            Pi[0][0]=Pi[0][0]+Xb                            
                        
                    else:
                        Pi.append([list(set(Xa + Xb)),tXab])
        if Pi:            
            charmRec(Pi,minsup,C)
        if not(isAContainedInAnyOfB(Xa,C)):             
            C.append(Xa) 
    
    C=C+list(orderP[:,0])
    return C
db=readTransactionalDatabase("shop.dat")
charm(db,1000)
```

```python
start=time()
charm(db,500)
finish=time()
totalCharm=start-finish
start=time()
genMax(db,500)
finish=time()
totalgenMax=start-finish
```
